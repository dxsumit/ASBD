## Question - 1
#### Suppose that the data for analysis includes the attribute age. The age values for the data tuples are (in increasing order) 13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70.

#### (a) Use min-max normalization to transform the values of age to the range[0:1].
#### (b) Use z-score normalization to transform the values of age.
#### (c) Use normalization by decimal scaling to transform the values of age such that the transformed value is less than 1.

```python
age = (13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70)

age = np.array(age)
mean = np.mean(age)
std = np.std(age)

min_val = age[0]
max_val = age[len(age)-1]
min_max_normal = []
z_score_normal = []
decimal_scale_normal = []

for i in range(len(age)):
    min_max_normal.append((age[i]-min_val)/(max_val-min_val))
    z_score_normal.append( (age[i]-mean)/std )
    decimal_scale_normal.append( age[i]/10**len(str(max_val)) )
    
    
print(min_max_normal)
print()
print(z_score_normal)
print()
print(decimal_scale_normal)
```

## Question - 2
### Dataset description 

```python
df = pd.read_csv("Avocado.csv")
df
```

```python
sorted_bin_means_df = df.sort_values(by="Total Volume", ascending=True)
volume = sorted_bin_means_df['Total Volume']

sorted_bin_means_df
```

### Bin Mean

```python
bin_size = volume.shape[0]//250
start = 0

for i in range(250):
    sorted_bin_means_df.iloc[start:start+bin_size, [2]] = [volume[start:start+bin_size].mean()] * bin_size
    start += bin_size
    
    
sorted_bin_means_df
```

### Bin Median

```python
sorted_bin_median_df = df.sort_values(by="Total Volume", ascending=True)
volume = df['Total Volume'].sort_values(ascending=True)

bin_size = volume.shape[0]//250
start = 0

for i in range(250):
    sorted_bin_median_df.iloc[start:start+bin_size, [2]] = [volume[start:start+bin_size].median()] * bin_size
    start += bin_size
    
    
    
sorted_bin_median_df
```

### bin-boundaries 

```python
sorted_bin_boundaries_df = df.sort_values(by="Total Volume", ascending=True)
volume = df['Total Volume'].sort_values(ascending=True)

bin_size = volume.shape[0]//250
start = 0

for i in range(250):
    currentBin = volume[start:start+bin_size]
    max_bin = currentBin.max()
    min_bin = currentBin.min()
    
    for i in range(len(currentBin)):
        if( (currentBin.iloc[i]-max_bin) > (currentBin.iloc[i]-min_bin) ):
            currentBin.iloc[i] = min_bin
        else:
            currentBin.iloc[i] = max_bin
    
    
    sorted_bin_boundaries_df.iloc[start:start+bin_size, [2]] = currentBin
    start += bin_size
    
    
    
sorted_bin_boundaries_df
```
